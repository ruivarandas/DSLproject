# Saliency Attribution Maps

```python
gen = present_values(folder, "saliency", return_data=True)
gen_Saliency = gen
correct = present(folder, "saliency", ['pred_results', 'ok'], but_zeros=False, return_data=True)
incorrect = present(folder, "saliency", ['pred_results', 'wrong'], but_zeros=False, return_data=True)
normal = present(folder, "saliency", ['true_labels', 'normal'], but_zeros=False, return_data=True)
abnormal = present(folder, "saliency", ['true_labels', 'abnormal'], but_zeros=False, return_data=True)
```

```python
fig, axs = plt.subplots(nrows=2, ncols=5, figsize=(10, 5), sharey=True, sharex=True)

fig.suptitle("Saliency Results", fontsize=16)

n_bins = 100

axs[0][0].hist(gen[0], bins=n_bins)
axs[0][0].set_xlabel('a')
axs[1][0].hist(gen[1], bins=n_bins)
        
axs[0][1].hist(normal[0], bins=n_bins)
axs[1][1].hist(normal[1], bins=n_bins)

axs[0][2].hist(abnormal[0], bins=n_bins)
axs[1][2].hist(abnormal[1], bins=n_bins)

axs[0][3].hist(correct[0], bins=n_bins)
axs[1][3].hist(correct[1], bins=n_bins)

axs[0][4].hist(incorrect[0], bins=n_bins)
axs[1][4].hist(incorrect[1], bins=n_bins)

for ax in axs:
    for a in ax:
        a.label_outer()

# plt.tight_layout()
plt.show()

plt.savefig('saliency.eps')
```
