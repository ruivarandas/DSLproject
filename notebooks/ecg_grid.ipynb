# Notebook to create figures with ECG grid

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import AutoMinorLocator
import numpy as np
from matplotlib.pyplot import figure
```

```python
def _ax_plot(secs=10):
    ax = plt.axes()
    ax.set_xticks(np.arange(0,11,0.2))    
    ax.set_yticks(np.arange(-2,3,0.5))
    ax.minorticks_on()
    ax.xaxis.set_minor_locator(AutoMinorLocator(5))
    ax.set_ylim(-1.8, 1.8)
    ax.set_xlim(0, secs)
    ax.grid(which='major', linestyle='-', linewidth='0.5', color='red')
    ax.grid(which='minor', linestyle='-', linewidth='0.5', color=(1, 0.7, 0.7))
    return ax
#     ax.plot(x,y, linewidth=0.5)
```

```python
x = [i for i in np.arange(0,10,0.2)]
y = [np.sin(i) for i in np.arange(0,10,0.2)]
```

```python
np.arange(0,11,0.2), np.arange(-2,3,0.5)
```

```python
fig = figure(1, figsize=(30,4.5))
_ = _ax_plot()
_ = plt.plot(x, y)
```
